Looking in more detail about manufacturer AIRBUS INDUSTRIE

```python
dfairbus = df.loc[df['manufacturer'] == 'AIRBUS INDUSTRIE']
dfairbus.head()
```

Getting number of  AIRBUS INDUSTRIE planes

```python
len(dfairbus.index)
```

Looking at planes numbers by year

```python
dfairbusmodel = dfairbus[['year','tailnum']]
dfairbusmodel.head()
```

```python
df_yeargroup = dfairbusmodel.groupby('year').count()
df_yeargroup.plot(kind = 'bar')
```
